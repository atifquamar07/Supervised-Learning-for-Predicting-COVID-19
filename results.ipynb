```python
import matplotlib.pyplot as plt
```

```python
# Names of models
models = ['k-NN', 'Random Forest', 'Multi-Layer Perceptron', 'Logistic Regression', 'SVM', 'XG-Boost']

# Accuracy scores for each model
accuracy_scores = [82.978, 80.843, 68.234, 66.523, 68.236, 79.586]  # Example scores

# Create a bar graph
plt.figure(figsize=(10, 6))  # Set figure size
plt.bar(models, accuracy_scores, color=['blue', 'green', 'red', 'yellow', 'magenta'])  # Bar plot

# Adding title and labels
plt.title('Raw Dataset Performance Metrics')
plt.xlabel('Models')
plt.ylabel('Accuracy')

plt.ylim(0, 100)

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Displaying the accuracy values on top of the bars
for i, score in enumerate(accuracy_scores):
    plt.text(i, score + 1, f'{score:.2f}', ha = 'center', va = 'bottom')  # Adjust vertical alignment for better readability

# Show the plot
plt.tight_layout()  # Adjust layout to prevent cutting off labels
plt.show()
```
